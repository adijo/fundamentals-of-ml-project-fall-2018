# Plot validation curves.


# CIFAR-10 Validation Curves
## Learning Rate


```python
lr = [0.1, 0.01, 0.001, 0.0001]
training_accuracy = [48, 90, 98, 97]
validation_accuracy = [42, 77.13, 80.92, 77.90]
plt.semilogx(lr, training_accuracy)
plt.semilogx(lr, validation_accuracy)
plt.xlabel('learning rate')
plt.ylabel('accuracy (%)')
plt.title("Validation Curve for CIFAR-10 with CNN")
plt.grid(True)
plt.legend(["training","validation"])
plt.savefig("CNN-CIFAR-10-Learning_Rate.png")
plt.show()
```

## Input Dropout


```python
dropout = [0, 5, 10, 15]
training_accuracy = [97, 96, 95, 95]
validation_accuracy = [81,79,79,77]

plt.plot(dropout, training_accuracy)
plt.plot(dropout, validation_accuracy)
plt.xlabel('Input Dropout (%)')
plt.ylabel('accuracy (%)')
plt.title("Validation Curve for CIFAR-10 with CNN")
plt.legend(["training","validation"])
plt.grid(True)
plt.savefig("CNN-CIFAR-10-Input_Dropout.png")
plt.show()
```

```python
## Weight Decay
decay= [0.1, 0.01, 0.001, 0.0001, 0]
training_accuracy = [62, 78, 89, 95, 98]
validation_accuracy = [61, 76, 80, 80, 81]

plt.semilogx(decay, training_accuracy)
plt.semilogx(decay, validation_accuracy)
plt.xlabel('Weight Decay')
plt.ylabel('Accuracy (%)')
plt.title("Validation Curve for CIFAR-10 with CNN")
plt.legend(["training","validation"])
plt.grid(True)
plt.savefig("CNN-CIFAR-10-Weight_Decay.png")
plt.show()
```

```python
## Transformations
decay= ["none","hflip", "hflip + rndcrop", "hflip+rndcrop+rotate"]
training_accuracy = [62, 78, 89, 95, 98]
validation_accuracy = [61, 76, 80, 80, 81]

plt.semilogx(decay, training_accuracy)
plt.semilogx(decay, validation_accuracy)
plt.xlabel('Weight Decay')
plt.ylabel('Accuracy (%)')
plt.title("Validation Curve for CIFAR-10 with CNN")
plt.legend(["training","validation"])
plt.grid(True)
plt.savefig("CNN-CIFAR-10-Weight_Decay.png")
plt.show()
```
